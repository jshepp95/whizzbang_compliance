```python
import os
import cv2
from ultralytics import YOLO
```

```python
# --- CONFIG
model_all = "/home/azureuser/projects/object_detection/models/yolo_11/yolo11n_ft_all0ww4b6z8.pt"
model_lozenge = "/home/azureuser/projects/object_detection/models/yolo_detector_lozenge/ft_lozenge.pt"

video_path = "/home/azureuser/projects/object_detection/archive/data/video/raw/vid_004.mp4"

video = "vid_004"
model = YOLO(model_lozenge)
```

```python
cap = cv2.VideoCapture(video_path, cv2.CAP_FFMPEG)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(f'detector_output_{video}.mp4', fourcc, fps, (frame_width, frame_height))

frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    print(f"Processing frame {frame_count}")

    results = model.predict(source=frame, save=False)[0]
    
    annotated_frame = results.plot()
    
    out.write(annotated_frame)

# Release resources
cap.release()
out.release()
```
