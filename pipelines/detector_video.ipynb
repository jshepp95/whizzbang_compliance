```python
import os
import cv2
from ultralytics import YOLO
```

```python
# --- CONFIG
# video_path = "/home/azureuser/projects/object_detection/archive/data/video/raw/vid_002.mp4"
video_path = "/home/azureuser/projects/object_detection/archive/data/video/raw/vid_003.mp4"
model_all = "/home/azureuser/projects/object_detection/models/yolo_11/yolo11n_ft_all0ww4b6z8.pt"
model_lozenge = "/home/azureuser/projects/object_detection/models/yolo_detector_lozenge/ft_lozenge.pt"

video = "vid_003"
model = YOLO(model_all)
```

```python
cap = cv2.VideoCapture(video_path, cv2.CAP_FFMPEG)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Check if video is valid
if not cap.isOpened():
    print("❌ Error: Could not open video file")
    exit()

# Set up Video Writer
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(f'detector_output_{video}.mp4', fourcc, fps, (frame_width, frame_height))

frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # End of video

    frame_count += 1
    print(f"Processing frame {frame_count}")

    # Run YOLO inference on frame
    results = model.predict(source=frame, save=False)[0]
    
    # Draw predictions on frame
    annotated_frame = results.plot()  # This returns the frame with boxes drawn
    
    # Write frame to output video
    out.write(annotated_frame)

# Release resources
cap.release()
out.release()
print("✅ Video processing complete!")
```
